Task1:

```python
from sklearn.datasets import make_moons
X, labels_true = make_moons(n_samples=500, noise=0.1)
```

KMEANS:

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
model.fit(X)
predY = model.labels_
```

```python
#original two moons dataset
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_true, cmap='viridis')
plt.title('Original Two Moons Data')


#clustering results from KMeans
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=predY, cmap='viridis')
plt.title('KMeans Clustering Results')

plt.show()
```

DBSCAN:

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.15, min_samples=5).fit(X)
labels = db.labels_
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_true, cmap='viridis')
plt.title('Original Two Moons Data')


plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN  Results')

plt.show()
```

GMM:

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=2)
gmm.fit(X)
y_cluster_gmm = gmm.predict(X)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_true, cmap='viridis')
plt.title('Original Two Moons Data')


plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_cluster_gmm, cmap='viridis')
plt.title('GMM  Results')

plt.show()
```

After visualizing the plots of the blobs dataset clustered using both K-Means and K-Medoids algorithms, it's evident that each algorithm behaves differently. K-Means seems to have created clusters with somewhat varying shapes, possibly due to its mean-based centroid calculation. On the other hand, K-Medoids appears to have produced more compact clusters, likely because it directly uses actual data points as centroids, enhancing its robustness to outliers.

K-Means is faster and suitable for globular clusters, while K-Medoids handles various cluster shapes and sizes better.

```python
from google.colab import files
from google.colab import files
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

uploaded = files.upload()
```

```python
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
pixels = image.reshape(-1, 3)

# Define the number of clusters
k_values = [2, 5, 10]

# Apply KMeans clustering for each value of k
plt.figure(figsize=(15, 5))
for i, k in enumerate(k_values):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)
    # Get the labels and cluster centers
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_

    # Replace each pixel with its corresponding cluster center
    segmented_image = centers[labels].reshape(image.shape)

    # Display the segmented image
    plt.subplot(2, len(k_values), i+1)
    plt.imshow(segmented_image.astype(np.uint8))
    plt.axis('off')
    plt.title(f'K={k}')

# Display the original image
plt.subplot(2, len(k_values), len(k_values)+1)
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')

plt.show()
```
